# Chapter 03 - Visualizing Data

```python
from matplotlib import pyplot as plt
```

## Linechart

```python
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2,543.3,1000,1111,15900,21255,391243]
```

```python
plt.plot(years, gdp,
        color = 'green',
        marker = 'o',
        linestyle = 'solid')
plt.title("Nominal GDP")
plt.ylabel("Billions of $")
plt.show()
```

## Bar Charts

```python
movies = ["Annie Hall", "Ben Hur","Casablance", "Ghandi","Avatar"]
num_oscars = [5,11,3,8,12]

plt.bar(
    range(len(movies)),
    num_oscars)

plt.title("Movies")
plt.ylabel("Acadamy Awards")
plt.xticks(
    range(len(movies)),
    movies)
```

### Histogram

```python
from collections import Counter
import random
```

```python
random.seed(10)
grades = [random.randrange(30,100) for _ in range(20)]
```

```python
# bucket in deciles (ranges of 10)
histogram = Counter(min(grade // 10 * 10,90) for grade in grades)
```

```python
plt.bar([x + 5 for x in histogram.keys()],
        histogram.values(),
        10, #height
        edgecolor=(0,0,0))

plt.xticks([10*i for i in range(11)])
plt.xlabel('Deciles')
plt.ylabel('# Students')
plt.show()
```

### Line Charts

```python
variance = [2**i for i in range(0,9)]
```

```python
bias = variance[::-1]
```

```python
total_error = [x+y for x,y in zip(variance,bias)]
```

```python
xs = [i for i,_ in enumerate(variance)]
```

```python
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias,'r-.',label='bias^2')
plt.plot(xs,total_error,'b:',label='total error')
plt.legend(loc = 0)
plt.xticks([])
plt.show()
```

# Scatter plots

```python
friends = [70,65,72,63,71,64,60,64,67]
minutes = [175,170,205,120,220,130,105,145,190]
```

```python
labels = ['a','b','c','d','e','f','g','h','i']
plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                xy = (friend_count,minute_count),
                xytext=(5,-5),
                textcoords = 'offset points')
```
